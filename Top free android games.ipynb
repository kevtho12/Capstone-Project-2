```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
game_data=pd.read_csv('Top Free Android Games.csv')
game_data.head()
```

```python
most_installs= game_data.groupby(['title','installs','total ratings','category'], as_index=False)
most_installs.count()[['title','installs','total ratings','category']].sort_values(['installs'], ascending=False).head(10)
```

```python
most_reviwed= game_data.groupby(['title','installs','total ratings','category'], as_index=False)
most_reviwed.count()[['title','installs','total ratings','category']].sort_values(['total ratings'], ascending=False).head(10)
```

```python
most_5stars= game_data.groupby(['title','installs','5 star ratings','category'], as_index=False)
most_5stars.count()[['title','installs','5 star ratings','category']].sort_values(['5 star ratings'], ascending=False).head(10)
```

```python
most_1stars= game_data.groupby(['title','installs','1 star ratings','category'], as_index=False)
most_1stars.count()[['title','installs','1 star ratings','category']].sort_values(['1 star ratings'], ascending=False).head(10)
```

```python
rating_game= game_data.groupby(['title','installs','average rating','category'], as_index=False)
rating_game.count()[['title','installs','average rating','category']].sort_values(['average rating'], ascending=False).head(10)
```

```python
lowrating_game= game_data.groupby(['title','installs','average rating','category'], as_index=False)
lowrating_game.count()[['title','installs','average rating','category']].sort_values(['average rating'], ascending=True).head(10)
```

```python
lowrating_game= game_data.groupby(['category'], as_index=True)
lowrating_game.mean()[['average rating']].sort_values(['average rating'], ascending=False)
```
